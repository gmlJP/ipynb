```python
import pandas as pd
import numpy as np
import csv as csv
import os

DIR_PATH= "/vagrant/titanic/"
```

```python
# TRAIN DATA
df = pd.read_csv(os.path.join(DIR_PATH, 'train.csv'), header=0)

# [              1st 2nd 3rd 
#  female ->  0  [0,   1,  2 , 
#  male   ->  1  [0,   1,  2]
# ]

median_ages = np.zeros((2,3)) # 2x3

# Add a Gender column (numeric)
df['Gender'] = df.Sex.map( {'female': 0, 'male' : 1} ).astype(int)

# Find median ages by Gender and Class
for i in range(0,2): # [0, 1]
    for j in range(0,3): # [0,1,2]
        median_ages[i,j] = df[(df.Gender == i) & (df.Pclass == j+1)]['Age'].dropna().median()
        
# Make a copy of Age
df['AgeFill'] = df['Age']

# Fill a new Age column with median age by Gender and Class
for i in range(0,2): # [0, 1]
    for j in range(0,3): # [0,1,2]
        df.loc[ (df.AgeFill.isnull()) & (df.Gender == i) & (df.Pclass == j+1) , 'AgeFill'] = median_ages[i,j]

df['FamilySize'] = df['SibSp'] + df['Parch']

df['Age*Class'] = df.AgeFill * df.Pclass

df = df.drop(['Age'], axis=1)

df = df.drop(['PassengerId', 'Name', 'Sex', 'Ticket', 'Cabin', 'Embarked'], axis=1)

train_df = df.copy()
```

```python
# TEST DATA
df = pd.read_csv(os.path.join(DIR_PATH, 'test.csv'), header=0)

# Add a Gender column (numeric)
df['Gender'] = df.Sex.map({'female': 0, 'male': 1}).astype(int)

# Make a copy of Age
df['AgeFill'] = df['Age']

# Fill a new Age column with median age by Gender and Class
for i in range(0,2): # [0, 1]
    for j in range(0,3): # [0,1,2]
        df.loc[ (df.AgeFill.isnull()) & (df.Gender == i) & (df.Pclass == j+1) , 'AgeFill'] = median_ages[i,j]
# Fare
#print df[df.Fare.isnull()]
df.loc[df.Fare.isnull(), 'Fare'] = df[df.Pclass ==3].Fare.median()
        
df['FamilySize'] = df['SibSp'] + df['Parch']

df['Age*Class'] = df.AgeFill * df.Pclass

# Save Id's
test_id_serie = df['PassengerId']

df = df.drop(['Age'], axis=1)

df = df.drop(['PassengerId', 'Name', 'Sex', 'Ticket', 'Cabin', 'Embarked'], axis=1)

test_df = df.copy()

test_df.info()

# TRAIN DATA
#Int64Index: 891 entries, 0 to 890
#Data columns (total 9 columns):
#Survived      891 non-null int64
#Pclass        891 non-null int64
#SibSp         891 non-null int64
#Parch         891 non-null int64
#Fare          891 non-null float64
#Gender        891 non-null int64
#AgeFill       891 non-null float64
#FamilySize    891 non-null int64
#Age*Class     891 non-null float64

# TEST DATA
#Pclass        418 non-null int64
#SibSp         418 non-null int64
#Parch         418 non-null int64
#Fare          418 non-null float64
#Gender        418 non-null int64
#AgeFill       418 non-null float64
#FamilySize    418 non-null int64
#Age*Class     418 non-null float64
```

```python
from sklearn.ensemble import RandomForestClassifier

# Numpy convert
train_data = train_df.values
test_data = test_df.values
ids = test_id_serie.values

forest = RandomForestClassifier(n_estimators = 100)

forest = forest.fit(train_data[0::, 1::], train_data[0::, 0])

output = forest.predict(test_data).astype(int)
```

```python
prediction_file = open(os.path.join(DIR_PATH,"myfirstforest.csv"), "wb")
open_file_object = csv.writer(prediction_file)
open_file_object.writerow(['PassengerId', 'Survived'])
open_file_object.writerows(zip(ids, output))
prediction_file.close()
```
