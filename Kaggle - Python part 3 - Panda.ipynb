```python
df['Gender'] = 4
df.head(5)
```

```python
df['Gender'] = df['Sex'].map( lambda x: x[0].upper() )
df.head(5)
```

```python
df.Gender = df.Sex.map( {'female': 0, 'male' : 1} ).astype(int)
df.head(5)
```

```python
# For the fun...
df['EmbarkedPlace'] = df.Embarked.map({'C': 'Cherbourg', 'Q': 'Queenstown', 'S': 'Southampton'}) # Queenstown, Cherbourg or Southampton
df.head(5)
```

```python
df.Age.dropna().mean()
```

```python
df.Age.dropna().median()
```

```python
median_ages = np.zeros((2,3)) # 2x3
median_ages
```

```python
for i in range(0,2): # [0, 1]
    for j in range(0,3): # [0,1,2]
        median_ages[i,j] = df[(df.Gender == i) & (df.Pclass == j+1)]['Age'].dropna().median()
```

```python
median_ages
```

```python
# Make a copy of Age
df['AgeFill'] = df['Age']
df.head()
```

```python
df[df.Age.isnull()][['Gender', 'Pclass', 'Age', 'AgeFill']].head(10)
```

[              1st 2nd 3rd

```python
for i in range(0,2): # [0, 1]
    for j in range(0,3): # [0,1,2]
        df.loc[ (df.AgeFill.isnull()) & (df.Gender == i) & (df.Pclass == j+1) , 'AgeFill'] = median_ages[i,j]
```

```python
df[df.Age.isnull()][['Gender', 'Pclass', 'Age', 'AgeFill']].head(10)
```
