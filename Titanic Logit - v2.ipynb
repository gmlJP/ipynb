```python
import pandas as pd
import numpy as np
import os
import re
import csv as csv
import statsmodels.api as sm
from statsmodels.nonparametric.kde import KDEUnivariate
from patsy import dmatrices

DIR_PATH= "/vagrant/titanic/"
```

```python
# TRAIN DATA
df = pd.read_csv(os.path.join(DIR_PATH, 'train.csv'), header=0)
```

```python
df.head()
```

```python
def search_title(name):
    match = re.search(r'\w+\.', name)
    if match:    
        return match.group()
    else:
        return ''
```

```python
# Create a new column for the title
df['Title'] = df['Name'].map(search_title)
```

```python
# Save the title
set_title = set(df['Title'])
```

```python
df.head()
```

```python
df_title = pd.DataFrame(columns=['Title', 'Mean Age', 'Median Age', 'Survived', 'Total', '%'])
[df.Pclass == 3]
for title in set_title:
    #pd_title = pd_title.append(pd.Series([title, df[df.Title == title].Age.mean(), df[df.Title == title].Age.median()]), ignore_index=True)
    df_title = df_title.append({'Title': title,
                                'Mean Age': df[df.Title == title].Age.mean(),
                                'Median Age' : df[df.Title == title].Age.median(),
                                'Survived': df[df.Title == title].Survived.sum(),
                                'Total': df[df.Title == title].Survived.count(),
                                '%': (df[df.Title == title].Survived.sum() / float(df[df.Title == title].Survived.count()))*100
                                #, 'Count' : df[df.Title == title].Title.value_counts()
                                }, ignore_index=True,
                               )
df_title.sort(columns=['%'], ascending=False)
df_title
```

```python
# Add a Gender column (numeric)
df['Gender'] = df.Sex.map( {'female': 0, 'male' : 1} ).astype(int)

# Make a copy of Age
df['AgeFill'] = df['Age']
```

```python
# Define the new column AgeFill with median age 
for title in set_title:
    df.loc[ (df.Age.isnull()) & (df.Title==title), 'AgeFill'] = df_title[ (df_title.Title==title) ]['Median Age'].iget(0)
```

```python
df = df.drop(['Age', 'Cabin', 'Ticket', 'Name', 'Embarked', 'Title', 'Sex'], axis=1)
```

```python
# Logit
dummy_ranks = pd.get_dummies(df.Pclass, prefix='Pclass')
dummy_ranks.head()
```

```python
cols_to_keep = ['Survived', 'Fare', 'Gender', 'AgeFill']
data = df[cols_to_keep].join(dummy_ranks.ix[:, 'Pclass_2':])
data['intercept'] = 1.0
data.head()
```

```python
train_cols = data.columns[1:]
train_cols
```

```python
logit = sm.Logit(data['Survived'], data[train_cols])

# fit the model
result = logit.fit()
```

```python
result.summary()
```

```python
result.conf_int()
```
