```python
from sklearn import tree
```

```python
X = [[10,20],
     [20,10],
     [15,40],
     [30,60],
     [10,88],
     [55,55],
     [100,88],
     ]
Y = [5,6,6,6,6,6,6]
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, Y)
```

```python
clf.predict([[2.,2.]])
```

```python
with open("test.dot", 'w') as f:
    f = tree.export_graphviz(clf, out_file=f, feature_names=['A', 'B'])
```

```python
! dot -Tpng test.dot -o test.png
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
frt = RandomForestClassifier(n_estimators=10)
frt = frt.fit(X, Y)
frt.predict([2.,2.])
```

```python
list_tree = enumerate([i for i in frt.estimators_])

for i, t in list_tree: 
#with open("test.dot", 'w') as f:
    f = open("forest_"+ str(i) +".dot", 'w')
    tree.export_graphviz(t, out_file=f, feature_names=['A', 'B'])
    f.close()
```
